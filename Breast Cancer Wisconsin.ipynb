```python
x=df.drop(["diagnosis"],axis=1)
y=df["diagnosis"]
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
xtrain.shape,xtest.shape
```

```python
xtrain.isnull().sum()
```

```python
xtest.isnull().sum()
```

```python
cols=xtrain.columns
```

```python
cols
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
xtrain = scaler.fit_transform(xtrain)
xtest = scaler.transform(xtest)
```

```python
xtrain = pd.DataFrame(xtrain,columns=[cols])
xtest = pd.DataFrame(xtest,columns=[cols])
```

```python
xtrain
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=9)
knn.fit(xtrain,ytrain)
```

```python
ypred = knn.predict(xtest)
```

```python
ypred
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(ytest,ypred))
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(ytest,ypred)
cm
```

```python
plt.figure(figsize=(10,6))
cm_matrix = pd.DataFrame(data=cm, columns=["True positive:1","True negative:0"],
                         index=["Prediction positive:1","Prediction negative:0"])
sns.heatmap(cm_matrix, annot=True, fmt="d", cmap="YlGnBu")
plt.show()
```

```python
scorelist=[]
for i in range (1,30):
    knn2 = KNeighborsClassifier(n_neighbors=i)
    knn2.fit(xtrain,ytrain)
    scorelist.append(knn2.score(xtest,ytest))
    
```

```python
plt.plot(range(1,30), scorelist)
plt.xlabel("Number of neighbors")
plt.ylabel("Accuracy Rate")
plt.show()
```
